## Analisis Korelasi

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

def analisis_korelasi(data):
    hasil = {}
    
    # Matriks korelasi Pearson
    hasil['pearson'] = data.corr(method='pearson')
    
    # P-value untuk korelasi Pearson
    p_values = pd.DataFrame(np.zeros_like(hasil['pearson']), 
                          columns=data.columns, 
                          index=data.columns)
    
    for i in range(len(data.columns)):
        for j in range(len(data.columns)):
            if i != j:
                corr, p_val = stats.pearsonr(data.iloc[:, i], data.iloc[:, j])
                p_values.iloc[i, j] = p_val
                
    hasil['p_values'] = p_values
    
    def buat_heatmap(matrix, title):
        plt.figure(figsize=(30, 20))
        sns.heatmap(matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
        plt.title(title)
        plt.tight_layout()
        return plt.gcf()
    
    # Visualisasi
    hasil['heatmap_pearson'] = buat_heatmap(hasil['pearson'], 'Matriks Korelasi Pearson')
    
    def interpretasi_korelasi(nilai):
        if abs(nilai) < 0.3:
            return "Lemah"
        elif abs(nilai) < 0.7:
            return "Sedang"
        else:
            return "Kuat"
    
    # Ringkasan interpretasi
    ringkasan = []
    for col1 in data.columns:
        for col2 in data.columns:
            if col1 < col2: 
                korelasi = hasil['pearson'].loc[col1, col2]
                p_value = hasil['p_values'].loc[col1, col2]
                interpretasi = interpretasi_korelasi(korelasi)
                signifikan = "signifikan" if p_value < 0.05 else "tidak signifikan"
                
                ringkasan.append({
                    'variabel_1': col1,
                    'variabel_2': col2,
                    'korelasi': korelasi,
                    'p_value': p_value,
                    'interpretasi': interpretasi,
                    'signifikansi': signifikan
                })
    
    hasil['ringkasan'] = pd.DataFrame(ringkasan)
    
    return hasil
```

```python
def get_kolom_korelasi_tinggi(df, threshold=0.8):
    korelasi = df.corr(method='pearson').abs()
    upper_tri = korelasi.where(np.triu(np.ones(korelasi.shape), k=1).astype(bool))
    
    # Mengidentifikasi kolom untuk dihapus
    kolom_drop = [kolom for kolom in upper_tri.columns 
                 if any(upper_tri[kolom] > threshold)]
    
    print(f"Kolom dengan korelasi tinggi yang bisa dihapus: {kolom_drop}")
    
    # Kolom yang tetap dipertahankan
    kolom_dipertahankan = [kolom for kolom in df.columns if kolom not in kolom_drop]
    return kolom_dipertahankan

# Mengambil kolom yang dipertahankan (yang tidak memiliki korelasi tinggi)
kolom_tidak_berkorelasi = get_kolom_korelasi_tinggi(train_df.drop(columns='Negara'), 0.9)
train_baru = train_df[kolom_tidak_berkorelasi]
train_baru
```

```python
import math
from xgboost import XGBRegressor
from sklearn.feature_selection import VarianceThreshold, mutual_info_regression
from sklearn.preprocessing import MinMaxScaler, FunctionTransformer, LabelEncoder
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
X_train = train_df.drop(columns=['Peningkatan Suhu Rata - Rata ° C'], axis=1)
y_train = train_df['Peningkatan Suhu Rata - Rata ° C']

scaler = RobustScaler()
X_train_scaled = scaler.fit_transform(X_train)

X_train_split, X_val, y_train_split, y_val = train_test_split(
    X_train_scaled, y_train, test_size=0.2, random_state=42
)
```

```python
model = XGBRegressor(
    n_estimators=200,
    learning_rate=0.01,  
    max_depth=3,       
    min_child_weight=3,  
    subsample=0.9,     
    colsample_bytree=0.9,
    gamma=0.1,          
    reg_alpha=0.1,      
    reg_lambda=1.0,      
    random_state=42
)

model.fit(
    X_train_split, 
    y_train_split,
    eval_set=[(X_val, y_val)],
    verbose=False
)
```

```python
y_pred_val = model.predict(X_val)
mse = mean_squared_error(y_val, y_pred_val)
r2 = r2_score(y_val, y_pred_val)
print(f"Validation MSE: {mse:.6f}")
print(f"Validation R²: {r2:.6f}")
```

```python
def mape(y_true: np.ndarray, y_pred: np.ndarray) -> float:
    mask = y_true != 0 
    return np.mean(np.abs(y_true[mask] - y_pred[mask]) / y_true[mask]) * 100
```

```python
print(f'MAPE Score: {mape(y_val, y_pred_val):.4f}%')
```
